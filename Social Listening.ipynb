```python
import pandas as pd

file_path_1 = "C:/Users/DINH KHANH VY/Desktop/Personal Project/KOL_KOC.xlsx"

sheets_dict_1 = pd.read_excel(file_path_1, sheet_name=None)

merged_df_kol = pd.DataFrame()

for sheet_name, df in sheets_dict_1.items():
    if sheet_name != 'Tiêu chí':
        df = df.rename(columns={
            'Link Fb': 'Link Facebook/Fanpage'
        })
        selected_col = df[['Tên', 'UID', 'Link Facebook/Fanpage', 'Nghề nghiệp','Phân loại']]
        selected_col = selected_col.rename(columns={
            'Tên': 'ten',
            'Link Facebook/Fanpage': 'link',
            'Nghề nghiệp': 'doi_tuong',
            'Phân loại': 'chu_de'
        })
        selected_col['chu_de'] = selected_col['chu_de'].replace({
            'Thời trang - Mỹ phẩm -  Làm đẹp': 'Thời trang - Mỹ phẩm - Làm đẹp'
        })
        merged_df_kol = pd.concat([merged_df_kol, selected_col], ignore_index=True)

merged_df_kol = merged_df_kol.drop_duplicates(subset='UID', keep='first')

print(merged_df_kol.head())
output_path = 'merged_df_kol_koc.csv'
merged_df_kol.to_csv(output_path, index=False, encoding='utf-8-sig')

print(f"Dữ liệu đã được lưu vào {output_path}")
```

```python
import pandas as pd

file_path_2 = "C:/Users/DINH KHANH VY/Desktop/Personal Project/Fanpage và group.xlsx"

# Đọc tất cả các sheet từ file Excel
sheets_dict_2 = pd.read_excel(file_path_2, sheet_name=None)

merged_df_fb = pd.DataFrame()

special_sheets = ['Thể thao', 'Công nghệ', 'Du lịch', 'Giao thông']

for sheet_name, df in sheets_dict_2.items():
    if sheet_name != 'Tiêu chí':
        if sheet_name == 'Tài chính' and None in df.columns:
            df.rename(columns={None: 'UID'}, inplace = True)
            
        df = df.rename(columns={
            'UID fb': 'UID',
            'UID facebook': 'UID',
            'UID FB': 'UID',
            'UID ': 'UID',
            'Unnamed: 3': 'UID'
        })
        if sheet_name in special_sheets:
            # Đổi tên cột 'Phân loại' thành 'Loại'
            if 'Phân loại' in df.columns:
                df.rename(columns={'Phân loại': 'Loại'}, inplace=True)
            # Thêm cột 'Chủ đề' với tên là sheet_name
            df['Chủ đề'] = sheet_name
        else:
            # Các sheet còn lại
            col_to_check = [
                'Loại Group ( công khai/kín)',
                'Loại Group (công khai/kín)',
                'Phân loại'
            ]
            for col in col_to_check:
                if col in df.columns:
                    df.rename(columns={col: 'Loại'}, inplace=True)
                    break
            # Đổi tên 'Phân loại' thành 'Chủ đề'
            if 'Phân loại' in df.columns:
                df.rename(columns={'Phân loại': 'Chủ đề'}, inplace=True)

                
        selected_col = df[['Tên fanpage/group ', 'UID', 'Link facebook', 'Loại', 'Chủ đề']]
        selected_col = selected_col.rename(columns={
            'Tên fanpage/group ': 'ten',
            'Link facebook': 'link',
            'Loại': 'doi_tuong',
            'Chủ đề': 'chu_de'
        })
        selected_col['doi_tuong'] = selected_col['doi_tuong'].str.strip().replace({
            'Kín': 'Group riêng tư',
            'Group kín': 'Group riêng tư',
            'Group  riêng tư': 'Group riêng tư',
            'Page': 'Fanpage',
            'fanpage': 'Fanpage',
            'Công khai': 'Group công khai'
        })
        
        merged_df_fb = pd.concat([merged_df_fb, selected_col], ignore_index=True)

merged_df_fb = merged_df_fb.drop_duplicates(subset='UID', keep='first')

print(merged_df_fb.head())
output_path = 'merged_df_fb_group.csv'
merged_df_fb.to_csv(output_path, index=False, encoding='utf-8-sig')

print(f"Dữ liệu đã được lưu vào {output_path}")
```
